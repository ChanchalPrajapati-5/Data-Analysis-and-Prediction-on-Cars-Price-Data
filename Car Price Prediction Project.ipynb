## Dealing with Missing Data

```python
# We will replace '?' with Nan using Numpy library. 
df.replace("?",np.NaN , inplace= True)
df.head()
```

Making function to count missing values in each column 

```python
missingvalues_check = df.isnull()
missingvalues_check.head()
```

We have following results:-

7. price column has 4 missing data

### REPLACEMENT OF MISSING VALUES


* Now as we have "Price" as targeted variable, we need to drop the rows with missing values as they are of no use in our analysis.

* "Num-of-doors" column has only two values or say two levels so we will replace missing values in that column by most frequent value.

* For rest of 5 columns we replace missing value by mean of values present in that column, as they have different numerical values.

### Dropping whole row with missing values in "price" column


```python
df.dropna(subset=["price"], axis=0, inplace=True)
#as we dropped 4 rows from our data
df.reset_index(drop=True, inplace=True)
```

```python
df
```

### Replacing missing values with most frequent value in column.

```python
df['num-of-doors'].value_counts()
```

As frequency of value '4' is much more than '2' so we will replace missing values in column "num-of-doors" by "four".

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
df
```

### Replacing missing value with mean for rest of the 5 columns.


```python
avg_norm = df["normalized-losses"].astype("float").mean(axis=0)
print("average normalised loss is" , avg_norm )
df["normalized-losses"].replace(np.nan, avg_norm , inplace=True)
df
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Bore column mean is" , avg_bore )
df["bore"].replace(np.nan, avg_bore , inplace=True)
df
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of stroke is " , avg_stroke )
df["stroke"].replace(np.nan, avg_stroke , inplace=True)
df
```

```python
avg_horsepower = df["horsepower"].astype("float").mean(axis=0)
print("Horsepower column mean is" , avg_horsepower )
df["horsepower"].replace(np.nan, avg_horsepower , inplace=True)
df
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
df
```

### Again Checking Columns For missing values.


```python
missingvalues_check = df.isnull()
missingvalues_check.head()
for column in missingvalues_check.columns.values.tolist():
    print(column)
    print (missingvalues_check[column].value_counts())
    print("")    
```

                          Now we have no missing values in our data.

# Data Formating

```python
df.dtypes
```


We should have columns of the correct data type.
Numerical variables - type 'float' or 'int'
Variables with strings such as categories should have type 'object'.

In our data we also have some incorrect data types so we have to convert data types into a proper format 
for each column using the "astype()" command.


```python
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

So we changed data type of 5 columns above we have correct data format.

```python
df.dtypes
```

When we do regression on data it does not understand words, so columns like fuel type and aspiration we need to introduce
dummy variables and add new columns depicting fuel and aspiration type as numbers.

```python
# change column names for clarity
dummy_var1.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_var1
```

```python
df
```

```python

# drop original column "aspiration" from "df"
df.drop("aspiration" , axis = 1, inplace=True)
```

```python
df = pd.concat([df, dummy_var1], axis=1)
```

```python
df
```

Similarly for fuel type column:-
    

```python
dummy_var2 = pd.get_dummies(df['fuel-type'])
dummy_var2.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_var2.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_var2], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df
```

 Now we have formatted data available with us.

# DATA NORMALIZATION

For 'length' , 'width' , 'height' column for normalisation , so that range of every value lise between 0 to 1 , we divide
every entry with maximum entry in that column.

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 

df[["length","width","height"]].head()
```

Now that we have data with no missing values and it is standardised and normalized , we save this clean data frame to again

## Multiple Linear Regression ( MLR ) Model 

From the previous section  we know that other good predictors of price could be:
Horsepower, Curb-weight, Engine-size ,Highway-mpg.
Let's develop a model using these variables as the predictor variables.


### MLR 1 - Using Horsepower, Curb-weight, Engine-size, Highway-mpg as predictor varibles in model.

```python
Z = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
```

```python
lm.fit(Z, df['price'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

Equation for predicted price using this model.




##### Price = (-15824.03802)+(53.6104 * Horsepower)+(4.7089 * Curb-weight)+(81.4722 * Engine-Size)+(36.3963 * Highway-Mpg)

```python
Y_hat = lm.predict(Z)
Y_hat
```

```python
ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

```python
Y_predict_multifit = lm.predict(Z)
```

```python
print('The mean square error of price and predicted value using multifit is: ', \
      mean_squared_error(df['price'], Y_predict_multifit))
```

```python
lm.fit(Z, df['price'])
# Find the R^2
print('The R-square is: ', lm.score(Z, df['price']))
```

So our MLR model can explain 80.94% variation in preditive variable "Price" ,which is a good R score.
